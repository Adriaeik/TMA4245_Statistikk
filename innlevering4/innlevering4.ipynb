Ola vil måle tyngdeakselerasjonen $g$ ved å slippe en stein utfor et høyt loddrett stup. Han holder steinen

neglisjerbar. Han lar $s$ betegne høyden på stupet og lar $t$ betegne tiden steinen bruker på fallet. 
Fra fysikkundervisningen på videregående skole husker han at sammenhengen mellom $s$, $g$ og $t$ er gitt som

\begin{align*}
s = \frac{1}{2}g t^2.
\end{align*}

Dersom han og Petter greier å måle $s$ og $t$ uten feil vil man dermed ha at 

\begin{align*}
g = \frac{2s}{t^2}.

Ola og Petter går nå på NTNU og vet at det er urealistisk å anta at de greier å måle $s$ og $t$ uten feil. De velger derfor å betrakte verdiene de måler for disse to størrelsene som realisasjoner av stokastiske variabler, og betegner disse stokastiske variablene med henholdsvis $S$ og $T$. Den tilhørende utregnede tyngdeakselerasjonen blir dermed en realisasjon av den stokastiske variabelen

\begin{align*}
G = \frac{2S}{T^2}.

Ut fra hvordan målingene av $s$ og $t$ gjøres anser de to gode vennene det naturlig å anta at $S$ og $T$ er
uavhengige stokastiske variabler. De lar dessuten $\mu_s$ og $\mu_t$ betegne de sanne (men ukjente) verdiene for henholdsvis $s$ og $t$, og antar at $E[S]=\mu_s$ og $E[T]=\mu_t$. Dessuten innfører de notasjonen 
$\sigma_s^2=Var[S]$ og $\sigma_t^2=Var[T]$.

### Deloppgave a)

Skriv pythonkode for å simulere $n=10\, 000$ realisasjoner av $G$, og bruk disse til å

Tror du $G$ er normalfordelt? For å undersøke dette: Lag et sannsynlighetshistogram av de $n$ genererte verdiene for $G$ og plott deretter (i samme plott som sannsynlighetshistogrammet) sannsynlighetstettheten til en normalfordeling med forventningsverdi og standardavvik lik de approksimerte verdiene for $\mu_g$ og $\sigma_g$.

$$Besvarelse$$

for å finne forventningsverdien til $G$, $\mu_g = E[G]$ og Variansen $\sigma_g^2 = Var[G]$. Brukar vi feilforplantningsformelen. Vi antar at $S$ og $T$ er uavh. og antar at $E[S]=\mu_s$ og $E[T]=\mu_t$. Dessuten innfører de notasjonen 
$\sigma_s^2=Var[S]$ og $\sigma_t^2=Var[T]$.

Forventnigsverdien $E[G]$ kan tilnærmast direkte fordi det er ein funksjon av uavh. stok. vatiabla ($S$ og $T$).
$$
\mu_g = E[G] = E[\frac{2S}{T}] = 2\frac{\mu_s}{\mu_t}=\underline{9.79}
$$

For variansen $\sigma_g^2 = Var[G]$ bruka vi feilforplantnigsformelen for funksjon av uavh. stok. var. 
\begin{align*}
\sigma_g^2 &=\left(\frac{\partial G}{\partial S}\right)^2\sigma_s^2 +\left(\frac{\partial G}{\partial T}\right)^2\sigma_t^2\\\\
&= \left(\frac{2}{\mu_t^2}\right)^2\sigma_s^2 + \left(-\frac{4\mu_s}{\mu_t^3}\right)^2\sigma_t^2\\\\
&= 7.79\\\\
\Rightarrow \sigma_g &= \sqrt{\sigma_g^2} = \underline{2.79}
\end{align*}


b) kodesnutt

```python
# Her kan du skrive pythonkoden din
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Definerer antall simuleringer og parametere
n = 10000
mu_s = 241.3  # målt verdi for s
mu_t = 7.02  # målt verdi for t
sigma_s = 2  # standardavvik for s
sigma_t = 1  # standardavvik for t

# Simulerer S og T basert på antatte normalfordelinger
S_sim = np.random.normal(mu_s, sigma_s, n)
T_sim = np.random.normal(mu_t, sigma_t, n)

# Beregner G basert på simulerte S og T
G_sim = 2 * S_sim / T_sim**2

# Beregner tilnærmede verdier for mu_g og sigma_g basert på simuleringene
mu_g_sim = np.mean(G_sim)
sigma_g_sim = np.std(G_sim)

# Skriver ut tilnærmede verdier
print(f"Tilnærmet mu_g basert på simulering: {mu_g_sim}")
print(f"Tilnærmet sigma_g basert på simulering: {sigma_g_sim}")

# Plotter sannsynlighetshistogram for G
plt.hist(G_sim, bins=50, density=True, alpha=0.6, color='g', label='Simulert G')

# Plotter sannsynlighetstettheten for en normalfordeling med tilnærmet mu_g og sigma_g
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mu_g_sim, sigma_g_sim)
plt.plot(x, p, 'k', linewidth=2, label='Normalfordeling')

plt.title("Sannsynlighetshistogram av G med normalfordeling")
plt.legend()
plt.show()
```

Kan du gi en teoretisk begrunnelse for at $\hat{\lambda}$ er (i hvert fall tilnærmet) normalfordelt?

* A)

SME for $\lambda$
\begin{align*}
L(\lambda) &= \prod_{i=1}^n  \frac{\lambda^{x_i}}{x_i!}e^{-\lambda}\\\\
&= \lambda^{\sum_{i=1}^n x_i}e^{-n\lambda}\prod_{i=1}^n  \frac{\lambda^{x_i}}{x_i!}
\end{align*}
For å maksimere sansynlightesestimatoren tar vi logaritmen av $L(\lambda)$ og deriverer for så å sette lik 0 for å finne toppunnkt, og bruker notasjonen $\mathcal{l}(\lambda) = ln(L(\lambda))$
\begin{align*}
\mathcal{l}(\lambda) &= \sum_{i=1}^n (x_i) ln(\lambda)-n\lambda +  ln\left(\prod_{i=1}^n\ \frac{\lambda^{x_i}}{x_i!}\right) \\\\
\mathcal{l}'(\lambda) &= \frac{\sum_{i=1}^nx_i}{\lambda} -n + 0\\\\
&\Rightarrow \lambda = \frac{\sum_{i=1}^nx_i}{n}
\end{align*}
Så SME for $\lambda$ blir $\hat{\lambda} = \frac{\sum_{i=1}^nx_i}{n}$

For å sjekke om $\hat{\lambda}$ er forventnigsrett skal vi finne forventningsverdien til  $\hat{\lambda}$

\begin{align*}
E[\hat{\lambda}] = E\left[\frac{\sum_{i=1}^nx_i}{n}\right] = \frac{1}{n}nE[x_i] = E[x_i] = \lambda
\end{align*} 
Forventnigsverdien blir $\lambda$ så då er den forventnigsrett

Variansen til $\hat{\lambda}$. i poissonfordeling er $Var(X)=\mu$

\begin{align*}
Var[\hat{\lambda}] = Var\left[\frac{\sum_{i=1}^nx_i}{n}\right] = \frac{1}{n^2}nVar[x_i] = \frac{1}{n}Var[x_i] = \frac{\lambda}{n}
\end{align*} 


* b)

For å finne eit utrykk for konstantane $\alpha$ og $\beta$ som gir ein best mulig estimator $\tilde{\lambda}$ må vi løyse følgande likning:
\begin{align*}
\tilde{\lambda} = \alpha\overline{X} + \beta\overline{Y} = \text{best mulig}\\\\
\end{align*}

vi har at $E[\overline{X}] = \lambda$ og $E[\overline{Y}] = \frac{1}{2}\lambda$ og forventngisverdien til $\tilde{\lambda}$ må då vere 
\begin{align*}
E[\tilde{\lambda}] = \alpha E[\overline{X}] + \beta E[\overline{Y}] = \lambda(\alpha + \frac{1}{2}\beta)\\
\end{align*}
Dersom $\tilde{\lambda}$ skal vere forventnigsrett må $E[\tilde{\lambda}] = \lambda$
\begin{align*}
\lambda = \lambda(\alpha + \frac{1}{2}\beta)\\
1 = \alpha + \frac{1}{2}\beta\\
\alpha + \frac{1}{2}\beta-1 = 0
\end{align*}

No løyser vi for Varians for å få enda ein liknig og så bruker vi Lagrange-regel for å minimere Variansen

Først berekner vi for $Var[\overline{X}] = \frac{\lambda}{n}$ og $Var[\overline{Y}] = \frac{\lambda}{2m}$
\begin{align*}
Var[\tilde{\lambda}] &=\alpha^2 Var[\overline{X}] + \beta^2 Var[\overline{Y}]\\
&= \alpha^2\frac{\lambda}{n} + \beta^2 \frac{\lambda}{2m}
\end{align*}


No kan vi bruke Lagrange-multiplikator med å innføre $L(\alpha,\beta,\gamma) = \alpha^2\frac{\lambda}{n} + \beta^2 \frac{\lambda}{2m} - \gamma(\alpha + \frac{1}{2}\beta-1)$. for så å partiellderivere $L(\alpha,\beta,\gamma)$ og sette opp eir system av likninga.

\begin{align}
L_{\alpha} &= 2\alpha\frac{\lambda}{n} - \gamma = 0\\
L_{\beta}  &= \beta\frac{\lambda}{m} - \frac{1}{2}\gamma = 0\\
L_{\gamma} &= \alpha +  \beta\frac{1}{2}-1 = 0 \\
\end{align}

Siden $\lambda$ er uinteressant kan vi la $\lambda$ "gå inn i" $\gamma$ kan vi gjere det enklare for oss 

\begin{align}
L_{\alpha} &= 2\alpha\frac{1}{n} - \gamma = 0\\
L_{\beta}  &= \beta\frac{1}{m} - \frac{1}{2}\gamma = 0\\
L_{\gamma} &= \alpha +  \beta\frac{1}{2}-1 = 0 \\
\end{align}

så dryssa vi på litt algebra

\begin{align*}
\gamma &= \alpha\frac{2}{n}  \ \tag{1}\\\\
\beta &= \frac{m}{2}\alpha\frac{2}{n} = \alpha\frac{m}{n} \tag{2}\\
\alpha &= \beta\frac{n}{m}\\\\
\beta\left(\frac{n}{m}+ \frac{1}{2} \right) &= 1  \tag{3}\\
\beta &= \frac{2m}{2n+m}\\\\
\alpha &= \frac{2m}{2n+m}\frac{n}{m}\tag{2}\\
\alpha &= \frac{2n}{2n+m}
\end{align*}

$\beta = \frac{2m}{2n+m} \ , \ \alpha = \frac{2n}{2n+m}$

```python
#Her kan du skrive kode for å simulere verdier og lage sannsynlighetshistogram
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt

# Definerer parameterverdier
n = 42
m = 35
lambda_real = 12
r = 100000

# Optimal verdi for alpha og beta basert på tidligere deloppgave
alpha_opt = 2*n / (2*n + m)
beta_opt = 2*m / (2*n + m)

# Initialiserer en tom liste for å lagre de simulerte verdiene av tilde_lambda
tilde_lambda_values = []

# Utfører r stokastiske simuleringer
for _ in range(r):
    # Simulerer X og Y verdiene
    X_values = np.random.poisson(lam=lambda_real, size=n)
    Y_values = np.random.poisson(lam=lambda_real/2, size=m)
    
    # Beregner estimatorene overline_X og overline_Y
    overline_X = np.mean(X_values)
    overline_Y = np.mean(Y_values)
    
    # Beregner estimatoren tilde_lambda basert på de simulerte verdiene
    tilde_lambda = alpha_opt * overline_X + beta_opt * overline_Y
    tilde_lambda_values.append(tilde_lambda)

# Plotter et sannsynlighetshistogram for de r simulerte verdiene for tilde_lambda
plt.hist(tilde_lambda_values, bins=25, density=True)
plt.title('Sannsynlighetshistogram for tilde_lambda')
plt.xlabel('tilde_lambda')
plt.ylabel('Sannsynlighettetthet')
plt.show()


# Generer qq-plot
sm.qqplot(np.array(tilde_lambda_values), line ='s')
plt.show()
```

Den er normalfordelt, sikkerhetsnivå type beinsikker

Forsøka er normalfordeltn $N(\mu,\ \sigma^2)$, så estimator for $\mu$ blir $\hat{\mu} = \overline{X} = \frac{1}{n}\sum_{i=1}^nX_i$
og $\overline{X}$ blir då normalfordelt $\overline{X} \ \sim \ N(\mu,\frac{\sigma^2}{n})$
det kan vi standarisere slik at $\frac{\overline{X}-\mu}{\sqrt{\frac{\sigma^2}{n}}}\ \sim \ N(0,1)$

Det gir oss uttrykket $P(-Z_{\frac{\alpha}{2}}\leq \frac{\overline{X}-\mu}{\sqrt{\frac{\sigma^2}{n}}} \leq Z_{\frac{\alpha}{2}}) = 1-\alpha$ 

no er det kun $\mu$ som er ukjent, og vi må løyse dei to ulikheitane med hensyn på $\mu$

\begin{align*}
-Z_{\frac{\alpha}{2}} &\leq \frac{\overline{X}-\mu}{\sqrt{\frac{\sigma^2}{n}}} \tag{1}\\
-Z_{\frac{\alpha}{2}}\frac{\sigma}{\sqrt{n}} &\leq \overline{X}-\mu\\
\mu &\leq \overline{X} + Z_{\frac{\alpha}{2}}\frac{\sigma}{\sqrt{n}}\\\\
\frac{\overline{X}-\mu}{\sqrt{\frac{\sigma^2}{n}}} &\leq Z_{\frac{\alpha}{2}} \tag{2}\\
\mu &\geq \overline{X} + Z_{\frac{\alpha}{2}}\frac{\sigma}{\sqrt{n}}
\end{align*}

Kofidensintervallet blir då
\begin{align*}
\left[\overline{X} - Z_{\frac{\alpha}{2}}\frac{\sigma}{\sqrt{n}} \ , \ \overline{X} + Z_{\frac{\alpha}{2}}\frac{\sigma}{\sqrt{n}}\right] = [\theta_L, \theta_U]
\end{align*}

Med $95\%$ konfidensintervall finner vi $Z_{\frac{\alpha}{2}}$ i tabell for $\frac{\alpha}{2} \Rightarrow Z_{\frac{\alpha}{2}} = 1.960$

### Deloppgave b)
Anta at temperaturene fra mars er uavhengige fra et år til et annet. Finn et $99\%$-konfidensintervall for forventet gjennomsnittstemperatur i mars.

$$Besvarelse$$

kjenner ikkje $\sigma$, så vi bruker $S$.  når vi bytter $\sigma$ med $S$ følger vi student t-fordeling, som har ein mindre friheitsgrad.

Vi har kartlagt at $X$ er normalfordelt.
Når vi har normalfordeling vil vi standarisere, men sidan $\sigma$ er ukjent  bytter vi $Z$ med $T$.

\begin{align*}
T &=  \frac{\overline{X}-\mu}{\sqrt{\frac{s^2}{n}}}\\\\
\overline{X} &= \frac{1}{n}\sum_{i=1}^nX_i\\\\
S^2 &= \frac{1}{n-1}\sum_{i=1}^n(X_i-\overline{X})^2
\end{align*}

Utleding av formelen for konfidensintervallet for t-fordeling.

Bruker vi Teoremet som sier: La $X_1,X_2,...,X_n$ vere tilfeldig utval fra $N(\mu,\sigma^2)$-populasjonen. Då er
$$V=\sum_{i=1}^n\left(\frac{X_i-\overline{X}}{\sigma}\right)^2 \sim \mathcal{X}_{n-1}^2$$ 
, og dessuten er $\overline{X}$ og $V$ uavh. stok. var.

merk at $V = \frac{(n-1)S^2}{\sigma^2}$. Ved å benytte dette får vi at 
\begin{align*}
T &=  \frac{\overline{X}-\mu}{\sqrt{\frac{s^2}{n}}} = \frac{\frac{\overline{X}-\mu}{\sqrt{\frac{\sigma^2}{n}}}}{\sqrt{\frac{\frac{(n-1)s^2}{\sigma^2}}{n-1}}} = \frac{Z}{\sqrt{\frac{V}{n-1}}} \sim t_{n-1} \\\\
\end{align*}

Altså er $T = t_{\frac{\alpha}{2}, n-1}$ student t-fordelt med $\nu = n-1$ (frihetsgrader). Så då kan vi bare løyse utrykket med hensyn på $\mu$.

\begin{align*}
T &=  \frac{\overline{X}-\mu}{\sqrt{\frac{s^2}{n}}}\\\\
T\frac{s}{\sqrt{n}} &= \overline{X}-\mu\\\\
\mu &= \overline{X} - t_{\frac{\alpha}{2}, n-1}\frac{s}{\sqrt{n}}
\end{align*}

Det gir intervallet
\begin{align*}
\left[\overline{X} - t_{\frac{\alpha}{2}, n-1}\frac{s}{\sqrt{n}} \ , \ \overline{X} + t_{\frac{\alpha}{2}, n-1}\frac{s}{\sqrt{n}}\right]
\end{align*}

Vi kan finne $t_{\frac{\alpha}{2}, n-1}$ i tabell : $\frac{\alpha}{2} = 0.005, n-1 = 11 \Rightarrow t_{\frac{\alpha}{2}, n-1} = 3.106$
og $s = \sqrt{\frac{1}{n-1}\sum_{i=1}^n(X_i-\overline{X})^2} =\sqrt{\frac{1}{n-1}}\sum_{i=1}^n(X_i-\overline{X})$ å berekne leddet $\sum_{i=1}^n(X_i-\overline{X})$ har eg brukt python til, fordi eg er lat. og fikk at $s = 2.526$ og $\overline{X} = 0.758$

Så intervallet blir då
$$(0.758\pm3.106\frac{2.526}{12})^\circ C = [-1.506 \ , \ 3.023]^\circ C$$

```python
# Beregninger og utskrift
n = len(x)  # Antall observasjoner
x_snitt = np.mean(x)  # Gjennomsnitt av X
s_x = np.std(x, ddof=1)  # Standardavviket til X med Bessel's korreksjon
alpha = 0.01  # For et 99% konfidensnivå
t_critical_x = stats.t.ppf(1 - alpha / 2, n - 1)  # Kritisk t-verdi for X
margin_of_error = t_critical_x * (s_x / np.sqrt(n))  # Margin of error
ci_lower = x_snitt - margin_of_error  # Nedre grense for konfidensintervallet
ci_upper = x_snitt + margin_of_error  # Øvre grense for konfidensintervallet

# Utskrift av resultater
print(f"Gjennomsnittstemperatur i mars (x_snitt): {x_snitt:.3f}°C")
print(f"Standardavviket til X (s_x): {s_x:.3f}")
print(f"Kritisk t-verdi for X: {t_critical_x:.3f}")
print(f"99% konfidensintervall for gjennomsnittstemperaturen i mars: ({ci_lower:.3f}, {ci_upper:.3f})°C")
```
