$Y_i = a x_i + \varepsilon_i$,

der $\varepsilon_1,\varepsilon_2,\ldots,\varepsilon_n$ antas uavhengige og identisk normalfordelt med forventningsverdi lik null og varians lik $\sigma^2$. Vi har dermed at $Y_1,Y_2,\ldots,Y_n$ er uavhengige stokastiske variabler, og $Y_i\sim N(ax_i,\sigma^2)$.

Merk at vi altså betrakter de observerte verdiene $y_1,y_2,\ldots,y_n$ som realisasjoner av stokastiske variabler $Y_1,Y_2,\ldots,Y_n$, mens verdiene $x_1,x_2,\ldots,x_n$ betrakter vi som kjente tall.

Modellen har to parametre, $a$ og $\sigma^2$, og vi ønsker å estimere verdien til disse fra de observerte parene $(x_1,y_1),(x_2,y_2),...,(x_n,y_n)$.

### Deloppgave a) **

Finn sannsynlighetsmaksimeringsestimatorene for $a$ og $\sigma^2$ og vis at disse kan skrives på formen

$$\hat{a}=\frac{\sum_{i=1}^n x_iY_i}{\sum_{i=1}^n x_i^2},\,\,\,\, 
\widehat{\sigma}^2 = \frac{1}{n}\sum_{i=1}^n \left(Y_i-\hat{a}x_i\right)^2.$$

#### Her er deloppgave b) slutt.

Det kan vises (NB: du trenger ikkje vise det) at 

$\sum_{i=1}^n \left(\frac{Y_i-\hat{a}x_i}{\sigma}\right)^2 \sim \chi^2_{n-1}$.

Dette resultatet kan du benytte til å besvare spørsmålene under.

### Deloppgave c) *

Finn forventningsverdien til $\widehat{\sigma}^2$. 

Forklar hvordan du kan se at $\widehat{\sigma}^2$ er forventningsskjev.

Foreslå en "korrigert" estimator for $\sigma^2$ (kall denne $\tilde\sigma^2$) som er forventningsrett.

Finn variansen til den forventningsrette estimatoren for $\sigma^2$.

$$Besvarelse$$

a) $Y_i\sim N(ax_i,\sigma^2)$

Rimlighetsfunksjon blir som følger

\begin{align*}
L(a,\sigma^2) &= \prod_{i=1}^n\left(\frac{1}{\sqrt{2\pi\sigma^2}}exp\left(-\frac{1}{2}\frac{(Y_i-ax_i)^2}{\sigma^2}\right) \right)\\
\mathcal{l}(a,\sigma^2) &= \sum_{i=1}^n\left(-\frac{1}{2}ln(2\pi)-\frac{1}{2}ln(\sigma^2) - \frac{1}{2\sigma^2}(Y_i-ax_i)^2\right)\\
&= -\frac{n}{2}ln(2\pi)-\frac{n}{2}ln(\sigma^2) - \frac{1}{2\sigma^2}\sum_{i=1}^n(Y_i-ax_i)^2\\
\end{align*}

Vi må så pariellderivere for og løyse for når utrykka er 0 for å finne sannsynlighetsmaksimeringsestimatorene for $a$ og $\sigma^2$.

\begin{align*}
\frac{\partial\mathcal{l}}{\partial a} &= - \frac{1}{2\sigma^2}\sum_{i=1}^n2(Y_i-ax_i)(-1)\\
&= \frac{1}{\sigma^2}\sum_{i=1}^n(Y_i-ax_i)\\
&= \frac{1}{\sigma^2}\left(\sum_{i=1}^nY_i-\sum_{i=1}^n ax_i\right) \tag{1}\\\\
\frac{\partial\mathcal{l}}{\partial \sigma^2} &= -\frac{n}{2\sigma^2} + \frac{1}{2\sigma^4}\sum_{i=1}^n(Y_i-ax_i)^2\\
&= \frac{1}{2\sigma^2}\left( -n + \frac{1}{\sigma^2}\sum_{i=1}^n(Y_i-ax_i)^2
\right) \tag{2}
\end{align*}

Vi må so sette $(1)$ og $(2)$ lik 0 for å finne sannsynlighetsmaksimeringsestimatorene.

\begin{align*}
a\sum x_i &= \sum Y_i \tag{1} \\
a &= \frac{\sum Y_i}{\sum x_i}\\
\Rightarrow \hat{a} &= \frac{\sum x_i Y_i}{\sum x_i^2}\\\\
\frac{1}{\sigma}^2\sum_{i=1}^n(Y_i-ax_i)^2 &= n \tag{2}\\
\sum_{i=1}^n(Y_i-ax_i)^2 &= n\sigma^2\\
\Rightarrow \hat{\sigma^2} &= \frac{1}{n}\sum_{i=1}^n(Y_i-ax_i)^2
\end{align*}

b)
\begin{align*}
E[\hat{a}] &= E\left[\frac{\sum x_i Y_i}{\sum x_i^2}\right]\\\\
&= \frac{1}{\sum x_i^2}E\left[\sum x_i Y_i\right]\\\\
&= \frac{\sum x_i}{\sum x_i^2}E\left[\sum Y_i\right]\\\\
&= \frac{1}{\sum x_i}a\sum x_i\\
&= a,\quad \text{forventningsrett } \ddot{\cup}
\end{align*}

\begin{align*}
Var[\hat{a}] &=Var\left[\frac{\sum Y_i}{\sum x_i}\right]\\\\
&= \frac{1}{\left(\sum x_i\right)^2}\sum Var\left[Y_i\right]\\\\
&= \frac{\sigma^2}{\sum_{i=1}^n x_i^2}
\end{align*}

c)
$\sum_{i=1}^n \left(\frac{Y_i-\hat{a}x_i}{\sigma}\right)^2 \sim \chi^2_{n-1}$.

\begin{align*}
E[ \chi^2_{n-1}] &= n-1\\
E\left[\sum_{i=1}^n\left(\frac{Y_i-\hat{a}x_i}{\sigma}\right)^2 \right]&= n-1\\
E\left[\frac{1}{\sigma^2} n\hat{\sigma^2} \right]&= n-1 \quad \text{fordi }\widehat{\sigma}^2 = \frac{1}{n}\sum_{i=1}^n \left(Y_i-\hat{a}x_i\right)^2. \\
E[\widehat{\sigma}^2] &= \frac{\sigma^2(n-1)}{n}\\
\end{align*}

Som vi ser er ikkje dette forventningsrett i hennhald til $ \chi^2$-fordeling 
Forå korigere dette definerer vi 
$$\tilde{\sigma}^2 = \frac{n}{n-1}\widehat{\sigma}^2$$
forventningsverdien blir da
\begin{align*}
E\left[ \tilde{\sigma}^2  \right] =  \frac{n}{n-2} E\left[\widehat{\sigma}^2 \right] = \frac{n}{n-1}\frac{\sigma^2(n-1)}{n} = \sigma^2
\end{align*}

Som er forventningsrett :)

For å finne variansen tar vi i bruk den nye estimatoren $\tilde{\sigma}^2$
Vi veit og at $Var[x] = 2\nu\Rightarrow Var\left[\chi^2_{n-1}\right] = \frac{1}{n^2} 2(n-1)$
\begin{align*}
Var\left[ \tilde{\sigma}^2  \right] &=  Var\left[\frac{n}{n-2} \widehat{\sigma}^2 \right] \\\\
&=\frac{n^2}{(n-2)^2} Var\left[ \frac{\sigma^2(n-1)}{n} \right]\\\\
&= \frac{n^2}{(n-2)^2}\sigma^4 Var\left[ \frac{\chi^2_{n-1}}{n} \right]\\\\
&= \frac{\sigma^4 n^2 2(n-1)}{(n-2)^2}\\\\
Var\left[ \tilde{\sigma}^2  \right] &= \frac{2\sigma^4}{n-1}
\end{align*}

I denne oppgaven skal du benytte stokastisk simulering til å utforske hvordan et residualplott ser ut når modellen som antas i enkel lineær regresjon er korrekt og hvordan residualplott ser ut i noen tilfeller hvor den antatte modellen ikkje er korrekt.

Vi skal starte med å anta følgende modell. For $i=1,2,\ldots,n$ la

$Y_i = 0.5 + 0.25 x_i + \varepsilon_i$,

der $\varepsilon_1,\varepsilon_2,\ldots,\varepsilon_n$ er uavhengige og normalfordelte med forventningsverdi lik null og varians lik $0.25^2$. Her er altså modellen som antas i enkel lineær regresjon korrekt, og parameterverdiene er $\alpha=0.5$, $\beta=0.25$ og $\sigma=0.25$.

I python-koden under har du fått oppgitt verdier for $x_i$ $i=1,2,...,25$. Deretter genereres tilhørende verdier for $y_1,y_2,\ldots,y_n$ ifølge modellen formulert ovenfor. De genererte verdiene visuliseres så i et spredningsplott.

```python
# Du trenger ikkje endre noe i denne koden!

import numpy as np
#from scipy.stats import norm
import matplotlib.pyplot as plt

#Initialisering av parameterverdier
n = 25
alpha = 0.5
beta = 0.25
sigma = 0.25

#Simulering av data etter modell
# x_1,x_2,...,x_n i intervallet [-5,5]
x = np.array([-3.842, -3.784, -3.745, -3.708, -3.37,  -3.288, -2.312, -2.078, -2.019, 
              -1.595,-1.106, -0.352, -0.171,  1.166,  2.196,  2.538,  2.772,  3.186,  
              3.309,3.876, 4.2,    4.261,  4.337,  4.352,  4.359])
# genererer tilhørende verdier for y_1,y_2,...,y_n
y = alpha + beta * x + np.random.normal(loc=0,scale=sigma,size=n) 
y_r = alpha + beta *x

#Visualiserer resultatet i et plott
plt.plot(x,y,'bo')
plt.xlabel('x')
plt.ylabel('y')

plt.plot(x,y_r)
plt.show()
```

Under er det gitt en python-funksjon som tar vektorer $x$ og $y$ som input og regner ut estimatene $\hat{\alpha}$, $\hat{\beta}$ og $S^2$ i en enkel lineær regresjonsmodell. Dette er tilsvarende metode som ble gjort i oppgave 1, bare nå med et konstantledd (SME for lineærregresjon, og forventningsrett estimator for variansen).

```python
# Du trenger ikkje endre noe i denne koden!

def estimerELR(x,y):
    #Beregner gjennomsnitt
    xStrek = np.mean(x)
    yStrek = np.mean(y)
    #Estimater for parametere
    betaHat = np.sum((x-xStrek)*y)/np.sum((x-xStrek)**2)
    alphaHat = yStrek - betaHat * xStrek
    S2 = np.sum((y-(alphaHat+betaHat*x))**2)/(len(x)-2)
    #Returnerer resultatet i en liste
    return [alphaHat,betaHat,S2]

paramHat = estimerELR(x,y)
print('alphaHat: ',paramHat[0])
print('betaHat: ',paramHat[1])
print('s2: ',paramHat[2])     
```

Ein liten samling av det vi har funne ut.
* $y = \alpha+\beta x +\epsilon \Rightarrow \hat{y}_0 =  \hat{\alpha} + \hat{\beta} x_0 + \epsilon$
* $\epsilon \sim N(0,\sigma^2)$
* $\hat{\beta} \sim N(\beta, \frac{\sigma^2}{S_{xx}}) = \frac{S_xy}{S_{xx}} $
* $\hat{\alpha} \sim N(\alpha,\sigma^2\frac{\sum x_i^2}{nS_{xx}}) = \overline{y} -\hat{\beta}\overline{x}$
* $\frac{\sum x_i^2}{nS_{xx}} = \frac{1}{n} + \frac{\overline{x}^2}{S_{xx}} $ fordi $\sum x_i^2 = n\overline{x}^2 + S_{xx}$
* Forventningsrett estimator for $\sigma^2$ er $S^2 = \frac{n\hat{\sigma^2}}{n-2} = \frac{1}{n-2}\sum(Y_i - (\hat{\alpha} + \hat{\beta}x_i))^2$,   $S^2$ er uavhenging av $\hat{\alpha}$ og $\hat{\beta}$
* $\frac{(n-2)s^2}{\sigma^2} \sim \chi_{n-2}^2$

Vidare finner ci at vi kan skrive 
\begin{align*}
\hat{y}_0 &=  \hat{\alpha} + \hat{\beta} x_0\\
Y_0 - \hat{y}_0 &= Y_0 - \hat{\alpha} - \hat{\beta} x_0 
\end{align*}

Sidan $\hat{alpha}$ og $\hat{\beta}$ er lineørfunksjoner av $Y_1,Y_2,..,Y_n$ bir differansen $Y_0 - \hat{Y}_0$ også ein lineærfunksjon av $Y_1,Y_2,..,Y_n \text{ og } Y_0$ og $Y_i$ er uavhengig og normalfirdelt.

Vidaare har vi då 

\begin{align*}
E[Y_0 - \hat{Y}_0] &= E[ Y_0 - \hat{\alpha} - \hat{\beta} x_0 ]\\
&= E[Y_0] - E[\hat{\alpha}] - x_0E[\hat{\beta}]\\
&= \alpha + x_0\beta - \alpha - x_0\beta = 0
\end{align*}
Altså er modellen forventnigsrett.

Vidare no bruker vi at $\hat{\alpha} = \overline{Y} - \overline{x}\hat{\beta}$

\begin{align*}
Var[Y_0 - \hat{Y_0}] &= Var[Y_0-\overline{Y} +\hat{\beta}\overline{x}- \hat{\beta}x_0]\\
&= Var[Y_0] + Var[\overline{Y}] + (x_0-\overline{x})^2Var[\hat{\beta}]\\
&= \sigma^2 + \frac{\sigma^2}{n} +(x_0-\overline{x})^2\frac{\sigma^2}{S_{xx}}\\
&=  \sigma^2 \left(1 + \frac{1}{n} +\frac{(x_0-\overline{x})^2}{S_{xx}} \right)
\end{align*}

Vi har no vist at vi kan sei at $ Y_0 - \hat{Y}_0 \sim N \left(0, \sigma^2 \left(1 + \frac{1}{n} +\frac{(x_0-\overline{x})^2}{S_{xx}} \right) \right)$

Den tilnærma standariseringa blir da 

$$Z = \frac{Y_0 - \hat{Y}_0 }{\sqrt{\sigma^2 \left(1 + \frac{1}{n} +\frac{(x_0-\overline{x})^2}{S_{xx}} \right)}} \sim N(0,1)$$

$\sigma^2$ er ukjent, så vi bytter den til den forventnigsrette estimatoren S^2, det betyr at vi mister 2 frihetsgrader.

Det kkan vi sjå ved at $T$ kan skivas på formen $\frac{Z}{\sqrt{V/\nu}}, \nu = n-2$

$$T = \frac{\frac{Y_0 - \hat{Y}_0 }{\sqrt{\sigma^2 \left(1 + \frac{1}{n} +\frac{(x_0-\overline{x})^2}{S_{xx}} \right)}}}{\sqrt{\frac{\frac{(n-2)s^2}{\sigma^2}}{(n-2)}}}, \text{ der } \frac{(n-2)s^2}{\sigma^2} = V \sim \chi_{n-2}^2$$
Så ergo er $T \sim t_{n-2}$
vidare definere vi dei kritiske verdiane i testen. $t_{\alpha/2, n-2}$ og $t_{1 - \alpha/2, n-2}$ med $\alpha = 1-0.9 = 0.1$
og siden student t-fordelinga er symetrisk utnyttar vi at $t_{1 - \alpha/2, n-2} = -t_{\alpha/2, n-2}$
Og vi kan da løyse for ulikheten 
\begin{align*}
&P \left(-t_{\alpha/2, n-2} \leq T \leq t_{\alpha/2, n-2}\right) = 1-\alpha\\
\Rightarrow &P\left(\hat{Y}_0 -t_{\alpha/2, n-2}\sqrt{s^2 \left(1 + \frac{1}{n} +\frac{(x_0-\overline{x})^2}{S_{xx}} \right)}\leq Y_0 \leq \hat{Y}_0 +t_{\alpha/2, n-2}\sqrt{s^2 \left(1 + \frac{1}{n} +\frac{(x_0-\overline{x})^2}{S_{xx}} \right)} \right)
\end{align*}

Bruker så at $\hat{Y}_0 = \hat{\alpha} +\hat{\beta}x_0$ og $S^2 = \frac{n\hat{\sigma}^2}{n-2} $

Då blir prediksjonsinteravallet

$$\hat{\alpha} +\hat{\beta}x_0 \pm t_{\alpha/2, n-2}\sqrt{\frac{n\hat{\sigma}^2}{n-2}\left(1 + \frac{1}{n} +\frac{(x_0-\overline{x})^2}{S_{xx}} \right)} 







































```python
# Her kan du skrive python-kode for å gjøre beregningene du trenger for å besvare oppgaven
from scipy import stats
hat_alpha = param[0]
hat_beta = param[1]
s2 = param[3]
n = len(x)
x_bar = np.mean(x)
S_xx = np.sum((x-x_bar)**2)
alpha = 0.10
t_critical = stats.t.ppf(alpha/2, n-2)  # t-verdien for 90% CI og n-2 frihetsgrader

x0 = np.linspace(24, 70, 100)
y_pred = hat_alpha + hat_beta * x0
SE = np.sqrt(s2 * (1 + 1/n + ((x0 - x_bar)**2) / S_xx))
lower_bound = y_pred - t_critical * SE
upper_bound = y_pred + t_critical * SE

plt.figure(figsize=(10, 6))
plt.plot(x0, y_pred, label='Prediksjon $\hat{y}_0$')
plt.fill_between(x0, lower_bound, upper_bound, color='gray', alpha=0.2, label='90% Prediksjonsintervall')
plt.scatter(x0, y_pred, color='blue', s=10)  # Fiktivt spredningsplott
plt.title('Prediksjonsintervall for Janka-hardhet')
plt.xlabel('Tetthet $x_0$')
plt.ylabel('Janka-hardhet $Y_0$')
plt.legend()
plt.show()
abs(np.mean(SE))
```


## Analyse av en RC-krets med Kondensator Koblet Parallelt med en Last


Når kondensatoren $ C $ er koblet parallelt med en lastmotstand $ R_L $ og spenningen måles over denne parallellkoblingen i en RC-krets, endres dynamikken i kretsen. Hovedkretsen består av en seriemotstand $ R $ og en parallellkrets av $ C $ og $ R_L $.


Impedansen til kondensatoren $ C $ er gitt ved $ \frac{1}{sC} $, og når den er koblet parallelt med $ R_L $, blir den totale impedansen for denne delen av kretsen:
$$
Z_{\text{total}} = \left( \frac{1}{sC} + \frac{1}{R_L} \right)^{-1} = \frac{R_L}{1 + sR_LC}
$$

Inkluderer vi seriemotstanden $ R $, blir den totale impedansen i kretsen $ R + Z_{\text{total}} $, og spenningen over parallellkoblingen av $ C $ og $ R_L $ (som er utgangsspenningen $ V_{out} $) kan uttrykkes ved spenningsdelingsprinsippet:
$$
V_{out} = V_{in} \times \frac{Z_{\text{total}}}{R + Z_{\text{total}}} = V_{in} \times \frac{\frac{R_L}{1 + sR_LC}}{R + \frac{R_L}{1 + sR_LC}}
$$

Som forenklast til $$ \frac{V_{out}}{ V_{in}} = \frac{R_L}{R(1 + sR_LC)+R_L}$$

* Frekvensrespons

Ved lave frekvenser ($ s $ nær 0), blir termen $ sR_LC $ ubetydelig sammenlignet med 1, så:
$$
V_{out} \approx V_{in} \times \frac{R_L}{R + R_L}
$$

Dette viser at utgangsspenningen er en del av inngangsspenningen basert på forholdet mellom $ R_L $ og $ R + R_L $, som kan være mindre enn 1 (eller i dB, mindre enn 0 dB), noe som indikerer et signal tap ved lave frekvenser.

Altså kan du aldri oppnå ein gain på meir enn $1$, og uansett kor liten $R_L$ blir vil gainen aldri bli heilt $1$ heller.

Ved høye frekvenser blir $ sR_LC $ stor, og:
$$
V_{out} \approx V_{in} \times \frac{R_L}{\infin} = 0
$$
Dette fører til lignende resultater som ved lave frekvenser, men viser spesielt at utgangsspenningen alltid er en fraksjon av inngangsspenningen ved alle frekvenser, potensielt alltid under 0 dB hvis $ R_L $ er tilstrekkelig lav sammenlignet med $ R $. Dette er grunnen til at det observeres negativ dB gain i dette oppsettet.
